```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = r"C:\Users\dariu\Desktop\Anomaly-detection-in-Microservices-based-systems\service anomalies dataset\service_anomalies.csv"
df = pd.read_csv(file_path)

# Vérifier les premières lignes du dataset
print(df.head())
```

```python
# Convertir la colonne 'IsError' en booléen si nécessaire
df['IsError'] = df['IsError'].astype(bool)

# Créer le pairplot avec distinction des anomalies
sns.pairplot(df, hue="IsError", palette={False: "blue", True: "red"})

# Afficher le graphe
plt.show()
```
